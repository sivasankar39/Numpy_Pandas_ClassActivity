Pandas

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/titanic_dataset.csv")
data
```

```python
data.head(10)
```

```python
percentage = data.groupby('Sex')['Survived'].mean() * 100
percentage
```

```python
mean_age = data['Age'].mean()
median_age = data['Age'].median()
std_age = data['Age'].std()

print(f"Mean:   {mean_age}")
print(f"Median: {median_age}")
print(f"Standard deviation:     {std_age}")
```

```python
no_of_passengers = data.groupby('Pclass').size()
no_of_passengers
```

```python
minor_females_survived = data[(data['Sex'] == 'female') & (data['Age'] < 18) & (data['Survived'] == 1)]
minor_females_survived
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data
#data['Age'].isna().sum()
```

```python
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

data['age_group'] = data['Age'].apply(categorize_age)
data
```

```python
data.nlargest(5, 'Fare')
```

```python
average_fare = data.groupby(['Pclass', 'Sex'])['Fare'].mean()
average_fare
```

```python
data = data.rename(columns={'Sex': 'Gender', 'Fare': 'Ticket_fare'})
data
```

```python
survival_by_class = data.groupby(['Pclass', 'Survived']).size()
survival_by_class
```

```python
average_age = data.groupby(['Pclass', 'Gender'])['Age'].mean()
average_age
```
